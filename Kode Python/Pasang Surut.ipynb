```python
#perhitungan muka air penting sesuai dengan tipe yang diperoleh
MLLW = 0 - (M2+K1+O1)
MHLW = 0 - abs(M2+(K1+O1))
MLHW = 0 + abs(M2-K1+O1)
MHHW = 0 + (M2+K1+O1)
print ("MLLW= ",MLLW)
print ("MHLW= ",MHLW)
print ("MLHW= ",MLHW)
print ("MHHW= ",MHHW)
```

```python
#Ganda
MLWS = 0 - (M2 + S2)
MLWN = 0 - abs(M2 - S2)
MHWN = 0 + abs(M2 - S2)
MHWS = 0 + (M2 + S2)
print ("MLWS= ",MLWS)
print ("MLWN= ",MLWN)
print ("MHWN= ",MHWN)
print ("MHWS= ",MHWS)
```
