```python
import os

import pandas as pd

paths = "/content/data"
files = os.listdir(paths)

df = pd.DataFrame()
for file in range(len(files)):
    if files[file].endswith('.xlsx'):
        # Read in File
        file_df = pd.read_excel(paths + "//" + files[file],
                                sheet_name="Data Harian - Table",
                                skiprows=[0,1,2,3,4,5,6,7,40,41,42,43,44,45,46,47,48])
        # Add to just this DataFrame
        # file_df['Month'] = str(files[file])
        # Update `df`
        df = df.append(file_df, ignore_index=True)
```

```python
clean3 = clean2.reset_index(level=None, drop=True, inplace=False, col_level=0, col_fill='')
clean3
```

```python
clean3.groupby(["ddd_car"]).count()
```

```python
clean4 = clean3[clean3["ddd_car"].str.contains("C") == False]
clean4.groupby(["ddd_car"]).count()
```

```python
clean4["ddd_car"] = clean4["ddd_car"].str.replace("NE","45")
clean4["ddd_car"] = clean4["ddd_car"].str.replace("NW","315")
clean4["ddd_car"] = clean4["ddd_car"].str.replace("SE","135")
clean4["ddd_car"] = clean4["ddd_car"].str.replace("SW","225")
clean4["ddd_car"] = clean4["ddd_car"].str.replace("E","90")
clean4["ddd_car"] = clean4["ddd_car"].str.replace("N","0")
clean4["ddd_car"] = clean4["ddd_car"].str.replace("S","180")
clean4["ddd_car"] = clean4["ddd_car"].str.replace("W","270")
clean4


```

```python
clean4['ddd_car'] = clean4['ddd_car'].astype('float')
clean4['Tanggal'] = clean4['Tanggal'].astype('str')
```

```python
hasil = clean4.groupby(['Tanggal']).mean()
hasil.to_csv('hasil.csv')
hasil2 = pd.read_csv('hasil.csv')
hasil2
```

```python
hasil2[['Tanggal1', 'Bulan', 'Kosong']] = hasil2["Tanggal"].apply(lambda x: pd.Series(str(x).split("-")))
hasil3 = hasil2.drop(['Tanggal','Kosong'],axis=1)
hasil3
```

```python
hasil4 = hasil3.sort_values(by='Bulan', ascending=True)
hasil5 = hasil4.reset_index(level=None, drop=True, inplace=False, col_level=0, col_fill='')
hasil5
```
